```python
from pathlib import Path
import re

p = Path("/content/Retrieval-based-Voice-Conversion-WebUI/infer/modules/vc/utils.py")
s = p.read_text(encoding="utf-8")

# Заменяем весь кривой блок try/except (с несколькими try) на один корректный.
pattern = r"import os,\s*pathlib\s*\n\s*\n(?:\s*try:\s*\n)+\s*from fairseq import checkpoint_utils\s*\n(?:\s*except Exception:\s*\n\s*checkpoint_utils\s*=\s*None\s*\n)+"
replacement = (
    "import os, pathlib\n\n"
    "try:\n"
    "    from fairseq import checkpoint_utils\n"
    "except Exception:\n"
    "    checkpoint_utils = None\n"
)

new_s, n = re.subn(pattern, replacement, s, count=1)
print("Replaced blocks:", n)

if n == 0:
    # если шаблон не совпал (бывает из-за пробелов), сделаем более мягкий вариант:
    new_s = re.sub(r"import os,\s*pathlib\s*\n", "import os, pathlib\n", s, count=1)
    # удаляем подряд идущие try/except вокруг checkpoint_utils и вставляем нормальный
    new_s = re.sub(
        r"(import os, pathlib\n)(?:.|\n)*?(?=def get_index_path_from_model)",
        "import os, pathlib\n\ntry:\n    from fairseq import checkpoint_utils\nexcept Exception:\n    checkpoint_utils = None\n\n",
        new_s,
        count=1
    )
    p.write_text(new_s, encoding="utf-8")
    print("Applied fallback patch.")
else:
    p.write_text(new_s, encoding="utf-8")
    print("Applied main patch.")

!sed -n '1,30p' /content/Retrieval-based-Voice-Conversion-WebUI/infer/modules/vc/utils.py
```

```python
%%bash
set -e

echo "=== ищем исходные wav ==="
find /content/dataset -type f \( -iname "*.wav" -o -iname "*.m4a" \)

echo
echo "=== создаём структуру ==="
mkdir -p /content/dataset/raw
mkdir -p /content/dataset/wavs

echo
echo "=== конвертация в правильный формат ==="

python - << 'PY'
import os, subprocess, glob, sys

raw_dir = "/content/dataset/raw"
out_dir = "/content/dataset/wavs"
os.makedirs(out_dir, exist_ok=True)

files = glob.glob(raw_dir+"/*.wav") + glob.glob(raw_dir+"/*.m4a")
if not files:
    print("❌ файлов в raw нет")
    sys.exit(1)

print("Найдено файлов:", files)

converted = []

for f in files:
    name = os.path.splitext(os.path.basename(f))[0]
    out = f"{out_dir}/{name}_40k.wav"

    cmd = [
        "ffmpeg","-y","-i",f,
        "-ac","1",
        "-ar","40000",
        "-vn",
        out
    ]
    subprocess.run(cmd, check=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    converted.append(out)

print("Готовые wav:", converted)

# ===== нарезка =====
import soundfile as sf
import numpy as np

slice_len = 10 * 40000   # 10 секунд
step = 8 * 40000        # перекрытие

final_dir = "/content/dataset/slices"
os.makedirs(final_dir, exist_ok=True)

total = 0

for f in converted:
    data, sr = sf.read(f)
    if len(data.shape)>1:
        data = data[:,0]

    pos = 0
    idx = 0
    while pos+sr*2 < len(data):   # минимум 2 секунды
        chunk = data[pos:pos+slice_len]
        out = f"{final_dir}/{os.path.basename(f)}_{idx}.wav"
        sf.write(out, chunk, sr)
        pos += step
        idx += 1
        total += 1

print("Нарезано кусков:", total)
print("Готово: /content/dataset/slices")
PY

echo
echo "=== проверим итог ==="
ls -lah /content/dataset/slices | head
```

```python
%%bash
set -e
FILE="/content/Retrieval-based-Voice-Conversion-WebUI/infer/lib/rmvpe.py"
BAK="${FILE}.bak"

# 1) восстановим из бэкапа (он у тебя есть, раз "PATCH already present")
if [ -f "$BAK" ]; then
  cp "$BAK" "$FILE"
  echo "Restored from $BAK"
else
  echo "ERROR: backup not found at $BAK"
  exit 1
fi

# 2) вставим корректный патч: оборачиваем конкретно '... ).to(device)' в try/except
python - <<'PY'
import re

path = "/content/Retrieval-based-Voice-Conversion-WebUI/infer/lib/rmvpe.py"
s = open(path, "r", encoding="utf-8").read()

marker = "RMVPE_TO_TRY_EXCEPT_PATCH"
if marker in s:
    print("Patch already present (marker found). Nothing to do.")
else:
    # Ищем ПЕРВОЕ вхождение ").to(device)" и заменяем на try/except блок.
    # Важно: сохраняем отступ строки.
    pattern = r"(?m)^(?P<indent>\s*)\)\.to\(device\)\s*$"
    m = re.search(pattern, s)
    if not m:
        raise SystemExit("Could not find a standalone ').to(device)' line to patch.")

    indent = m.group("indent")
    repl = (
        f"{indent}{marker} = True\n"
        f"{indent}try:\n"
        f"{indent}    _tmp = ).to(device)\n"
        f"{indent}except RuntimeError as e:\n"
        f"{indent}    if 'No CUDA GPUs are available' in str(e) or 'CUDA error' in str(e):\n"
        f"{indent}        print('RMVPE: CUDA failed → switching to CPU')\n"
        f"{indent}        device = 'cpu'\n"
        f"{indent}        _tmp = ).to(device)\n"
        f"{indent}    else:\n"
        f"{indent}        raise\n"
    )

    # Чтобы сохранить исходную структуру, заменяем только само окончание ").to(device)"
    # на блок, который возвращает _tmp; потом ниже заменим первое использование результата.
    s2 = re.sub(pattern, repl, s, count=1)

    # Теперь подменяем первое присваивание "... = (... ).to(device)" на "... = (... _tmp"
    # Мы не знаем имя переменной, поэтому делаем локальную замену: ближайшее " = " перед "_tmp"
    # Проще: заменим первое вхождение "_tmp = ).to(device)" на ").to(device)" в исходном контексте уже внутри try.
    # А дальше заменим сразу следующее ") .to(device)"-окончание в выражении на "_tmp".
    # Для этого: после вставки, первое появление строки "_tmp = ).to(device)" оставляем,
    # а первое появление самого текста ").to(device)" в выражении заменяем на ")"
    # и добавляем ниже строку, которая возвращает _tmp — слишком рискованно.

    # Надёжнее: вместо " _tmp = ).to(device)" мы делаем так:
    # 1) В try присваиваем __patched_obj = (<ОРИГИНАЛ_ВЫРАЖЕНИЕ>).to(device)
    # Но выражение у нас не захвачено.
    #
    # Поэтому делаем другой, куда более безопасный патч:
    # заменить " ).to(device)" на:
    #   ).to(device)
    #   <indent>...   # нельзя — ломает выражение
    #
    # => Откатываемся и делаем патч по месту в rmvpe.py вручную через более широкий regex:
    # найдём первый фрагмент " = ( ... ).to(device)" и обернём весь RHS в try/except.

    # Откатываемся к исходному и делаем широкий патч:
    s = open(path, "r", encoding="utf-8").read()
    pat2 = r"(?s)(?P<lhs>^\s*\w+\s*=\s*)(?P<rhs>\(.*?\)\.to\(device\))"
    m2 = re.search(pat2, s, re.M)
    if not m2:
        raise SystemExit("Could not find assignment with '( ... ).to(device)' to patch safely.")
    lhs, rhs = m2.group("lhs"), m2.group("rhs")
    indent2 = re.match(r"^(\s*)", lhs).group(1)

    block = (
        f"{indent2}{marker} = True\n"
        f"{indent2}try:\n"
        f"{indent2}    {lhs.strip()} {rhs}\n"
        f"{indent2}except RuntimeError as e:\n"
        f"{indent2}    if 'No CUDA GPUs are available' in str(e) or 'CUDA error' in str(e):\n"
        f"{indent2}        print('RMVPE: CUDA failed → switching to CPU')\n"
        f"{indent2}        device = 'cpu'\n"
        f"{indent2}        {lhs.strip()} {rhs.replace('device','device')}\n"
        f"{indent2}    else:\n"
        f"{indent2}        raise\n"
    )

    # Заменяем ровно первое вхождение этого присваивания
    s3 = s[:m2.start()] + block + s[m2.end():]

    open(path, "w", encoding="utf-8").write(s3)
    print("Patched successfully with try/except fallback.")
PY

# 3) проверим, что файл теперь хотя бы парсится питоном
python -m py_compile "$FILE"
echo "py_compile OK"

# 4) покажем 40 строк вокруг маркера, чтобы ты видел, что вставилось куда надо
python - <<'PY'
import re
p="/content/Retrieval-based-Voice-Conversion-WebUI/infer/lib/rmvpe.py"
s=open(p,"r",encoding="utf-8").read().splitlines()
idx=[i for i,l in enumerate(s) if "RMVPE_TO_TRY_EXCEPT_PATCH" in l]
if not idx:
    print("Marker not found")
else:
    i=idx[0]
    a=max(0,i-20); b=min(len(s), i+20)
    print("\n".join(s[a:b]))
PY
```
